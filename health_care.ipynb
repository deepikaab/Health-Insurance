```python
import pandas as pd
if not isinstance(corre, pd.DataFrame):
    corre = pd.DataFrame(corre)
threshold = 0.0

important_pairs = []
for i in range(len(corre.columns)):
    for j in range(i):
        if abs(corre.iloc[i, j]) > threshold:
            important_pairs.append((corre.columns[i], corre.columns[j], corre.iloc[i, j]))

# Extract unique important attributes
important_attributes = set()
for attr1, attr2, _ in important_pairs:
    important_attributes.add(attr1)
    important_attributes.add(attr2)

count_important_attributes = len(important_attributes)
count_important_attributes, important_attributes
```

```python

import matplotlib.pyplot as plt

# Create box plots for the important attributes
plt.figure(figsize=(20, 15))

for i, attribute in enumerate(important_attributes, 1):
    plt.subplot(len(important_attributes) // 3 + 1, 3, i)
    sns.boxplot(data=final_merged_data[attribute])
    plt.title(attribute)

plt.tight_layout()
plt.show()
```

```python
def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR  # Correct indentation here
    upper_bound = Q3 + 1.5 * IQR  # Correct indentation here
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return len(outliers)

# Count outliers for each important attribute
outlier_counts = {attribute: count_outliers(final_merged_data, attribute) for attribute in important_attributes}

# Display the number of outliers for each attribute
for attribute, count in outlier_counts.items():
    print(f"{attribute}: {count} outliers")
```

```python
# Define a threshold for the number of outliers
outlier_threshold = 6000  # Adjust this value based on your criteria

# Extract attributes with fewer outliers than the threshold
attributes_with_fewer_outliers = [attribute for attribute, count in outlier_counts.items() if count <= outlier_threshold]

print("Attributes with fewer outliers:")
print(attributes_with_fewer_outliers)
```

```python
plt.figure(figsize=(20, 15))

for i, attribute in enumerate(attributes_with_fewer_outliers, 1):
    plt.subplot(len(attributes_with_fewer_outliers) // 3 + 1, 3, i)
    sns.boxplot(data=final_merged_data[attribute])
    plt.title(attribute)

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the data
scaler = StandardScaler()
dd_scaled = scaler.fit_transform(final_merged_data[attributes_with_fewer_outliers])

# Convert back to a DataFrame
dd_scaled = pd.DataFrame(dd_scaled, columns=attributes_with_fewer_outliers)
```

```python
from sklearn.decomposition import PCA

# Apply PCA
pca = PCA()
pca.fit(dd_scaled)

# Get explained variance ratio
explained_variance = pca.explained_variance_ratio_
```

```python
import numpy as np
import pandas as pd

components = pca.components_

# Get the absolute values of the components
abs_components = np.abs(components)

# Sum the absolute values across components to get the overall importance of each feature
feature_importance = np.sum(abs_components, axis=0)

# Create a DataFrame to display the feature importance
# Use the column names from dd_scaled instead of the entire DataFrame
feature_importance_df = pd.DataFrame({
    'Attribute': dd_scaled.columns,  # Use column names here
    'Importance': feature_importance
})

# Sort by importance
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

print(feature_importance_df)
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(x='Importance', y='Attribute', data=feature_importance_df)
plt.title('Feature Importance Based on PCA')
plt.xlabel('Importance')
plt.ylabel('Attribute')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder

# Set the number of clusters
n_clusters = 3

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Iterate over columns and encode non-numerical ones
for col in dd_selected.columns:
    if dd_selected[col].dtype == 'object':
        dd_selected[col] = label_encoder.fit_transform(dd_selected[col])

# Initialize the KMeans model
kmeans = KMeans(n_clusters=n_clusters, random_state=42)

# Fit the KMeans model to the encoded data
kmeans.fit(dd_selected)

# Predicting the clusters
dd_selected['cluster'] = kmeans.labels_

# Displaying the dataset with cluster labels
print(dd_selected[['cluster']].head())
```

```python
# Assuming cluster 1 represents fraud and cluster 0 represents genuine
dd_selected['fraud_prediction'] = dd_selected['cluster'].apply(lambda x: 'Y' if x == 1 else 'N')

# Displaying the dataset with fraud predictions
print(dd_selected[['cluster', 'fraud_prediction']].head())
```

```python
# Display the count for 'Y' and 'N'
fraud_counts = dd_selected['fraud_prediction'].value_counts()
print(f"Counts for fraud ('Y') and genuine ('N'):\n{fraud_counts}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the percentage of 'Y' (fraud) and 'N' (genuine) predictions
fraud_percentage = dd_selected['fraud_prediction'].value_counts(normalize=True) * 100

# Convert to a DataFrame for easier plotting
fraud_percentage_df = fraud_percentage.reset_index()
fraud_percentage_df.columns = ['Prediction', 'Percentage']

# Plotting the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='Prediction', y='Percentage', data=fraud_percentage_df, palette='flare')

# Adding labels and title
plt.title('Percentage of Fraud (Y) and Genuine (N) Transactions')
plt.xlabel('Prediction')
plt.ylabel('Percentage (%)')

# Show the percentage on the bars
for index, row in fraud_percentage_df.iterrows():
    plt.text(index, row['Percentage'], f"{row['Percentage']:.2f}%", color='black', ha="center", va="bottom")

# Show the plot
plt.show()
```
